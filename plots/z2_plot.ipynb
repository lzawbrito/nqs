```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.serif'] = ['Computer Modern Sans']
plt.rcParams['text.latex.preamble'] = '\\usepackage[cm]{sfmath}'
plt.rcParams['font.family'] = 'sans-serif'
```

```python
data = np.loadtxt("../data/z2_varying_h_3x3.txt")
thetas, exact_energies, genn_energies, genn_energies_var = data[:, 0], data[:, 1], data[:, 2], data[:, 3]
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
plt.plot(thetas, genn_energies, color='black', linestyle='dashed', alpha=0.5, label='GENN')
ax.fill_between(thetas, genn_energies - genn_energies_var, genn_energies + genn_energies_var,
				 color='cornflowerblue', alpha=0.4)
ax.plot(thetas, exact_energies, color='cornflowerblue', alpha=0.8, label='exact')
ax.set_xlabel('$\\tan^{-1}g/K$')
ax.set_ylabel('Energy')
ax.legend()
ax.set_title('$3\\times 3$ lattice, 1000 VMC samples, 500 epochs')
plt.savefig('z2_varying_h_3x3.pdf')
```
